You have been asked to support a team of researchers who have been collecting data about penguins in Antartica! 

**Origin of this data** : Data were collected and made available by Dr. Kristen Gorman and the Palmer Station, Antarctica LTER, a member of the Long Term Ecological Research Network.

**The dataset consists of 5 columns.**

- culmen_length_mm: culmen length (mm)
- culmen_depth_mm: culmen depth (mm)
- flipper_length_mm: flipper length (mm)
- body_mass_g: body mass (g)
- sex: penguin sex

Unfortunately, they have not been able to record the species of penguin, but they know that there are three species that are native to the region: **Adelie**, **Chinstrap**, and **Gentoo**, so your task is to apply your data science skills to help them identify groups in the dataset!

```python
# Import Required Packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

## Loading and examining the dataset

```python
df = pd.read_csv("data/penguins.csv")
penguins_df=df
df=df.dropna()
columns=['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm','body_mass_g']
Q1,Q3=np.quantile(df['flipper_length_mm'],.25),np.quantile(df['flipper_length_mm'],.75)
iqr=Q3-Q1
df=df[(df['flipper_length_mm']>=Q1-1.5*iqr)&(df['flipper_length_mm']<=Q3+1.5*iqr)]
for c in columns:
    sns.boxplot(data=df,x=c)
    plt.show()
```

## Null values and outliers removed.

## Preprocessing

```python
penguins_clean=df
df=pd.get_dummies(df,drop_first=True)
scaler=StandardScaler()
df=scaler.fit_transform(df)
penguins_preprocessed=df
df
```

## Intrinsic => n_components

```python
pca=PCA()
pca.fit(df)
transformed=pca.transform(df)
features=range(pca.n_components_)
plt.bar(features,pca.explained_variance_ratio_)
n_components=2
```

## Actual PCA

```python
pca=PCA(n_components=2)
df=penguins_preprocessed
transformed=pca.fit_transform(df)
penguins_PCA=pd.DataFrame(transformed)
df=penguins_PCA
penguins_PCA
```

## Elbow => n_cluster

```python
import warnings
warnings.simplefilter('ignore')
inertias=[]
for i in range(1,10):
    model=KMeans(n_clusters=i)
    model.fit_predict(df)
    inertias.append(model.inertia_)
plt.plot(range(1,10),inertias)
plt.show()
n_clusters=4
```

## Actual Kmeans

```python
kmeans=KMeans(n_clusters=n_clusters, random_state=42).fit(penguins_PCA)
penguins_clean['label']=kmeans.labels_
df=penguins_clean
stat_penguins=df.groupby('label')[columns].mean()
stat_penguins
```

```python
df=penguins_PCA
df.columns=['A','b']
sns.scatterplot(data=df,x='A',y='b',c=kmeans.labels_)
plt.show()
```
